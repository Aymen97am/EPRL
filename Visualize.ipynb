```python
import sys 
sys.path.append("/home/aalmarja/.local/lib/python3.9/site-packages")
import numpy as np 
import pandas as pd
import matplotlib.pyplot as plt
import pickle
import os
import seaborn as sns
sns.set_style("whitegrid")
```

```python
np.arange(0.05, 2.5, 0.1).shape
```

```python
with open(f"./results/logs/processed/DATA.pkl","rb") as file:
            DATA = pickle.load(file)
```

```python
df = DATA[(DATA["maxD"] > DATA["AmaxCov"]) & (DATA["AmaxCov"] < DATA["BPI_UCRL"])]
```

```python
df[df["alpha"]< 2]
```

```python
df["perf"] = df["maxD"]/df["BPI_UCRL"]
df[df["perf"] == df["perf"].max()]
```

```python
#MDP where BPI_UCRL is best: d5854bc4-0544-4ff8-bfa4-a419c2d703d5
#MDP where maxD is best: 4d51ec38-7def-47f4-a574-4b76783cfbd5
#MDP where maxCov is best: 1b1bbfec-03e6-4d26-a220-4cf268107d2e
file = open("./results/logs/S2A2H3_1b1bbfec-03e6-4d26-a220-4cf268107d2e.pkl", "rb")
df, correct, minGap = pickle.load(file)
n_sim = 10
algos = ["BPI_UCRL", "maxD", "AmaxCov"]
diameter = {"BPI_UCRL" : "Qstar_diameter", "maxD": "max_diameter", "AmaxCov": "max_diameter"}
epsilon = 1.55*minGap
data = {"Algorithm" : [], "tau": []}
for name in algos:
    df_alg = df[(df["name"]== name)]
    for sim in range(n_sim):
        df_alg_sim = df_alg[df_alg["n_simu"]== sim]
        max_ep = df_alg_sim["episode"].max()
        df_alg_sim = df_alg_sim[(df_alg_sim[diameter[name]] < epsilon)]
        tau = df_alg_sim["episode"].min()
        if np.isnan(tau): 
            tau = max_ep
        data["tau"].append(tau)
        if name == "AmaxCov":
            data["Algorithm"].append("maxCov")
        else:
            data["Algorithm"].append(name)
```

```python
dd = pd.DataFrame(data)
```

```python
matplotlib.rc('xtick', labelsize=20) 
matplotlib.rc('ytick', labelsize=20)
f, ax = plt.subplots(figsize=(7, 6))
ax = sns.boxplot(x="Algorithm", y="tau", data=dd,
            whis=[5, 95], width=.4, linewidth = 1.5)
ax = sns.swarmplot(x="Algorithm", y="tau", data=dd, color=".25")
tau1 = np.mean(dd[dd["Algorithm"]=="BPI_UCRL"]["tau"])
tau2 = np.mean(dd[dd["Algorithm"]=="maxD"]["tau"])
tau3 = np.mean(dd[dd["Algorithm"]=="maxCov"]["tau"])
plt.axhline(y=tau1, color="blue", linestyle="--", linewidth = 1.9)
plt.axhline(y=tau2, color="orange", linestyle="--", linewidth = 1.9)
plt.axhline(y=tau3, color="green", linestyle="--", linewidth = 1.9)
ax.yaxis.grid(True)
ax.set_ylabel("")
ax.set_xlabel("")
#ax.legend()
sns.set_context("paper")
sns.despine(trim=False, left=False)
plt.savefig("maxCov.pdf")
```
